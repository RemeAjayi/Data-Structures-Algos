```python
class Hashtable():
    def __init__(self):
        self.MAX = 100
        self.arr = [None for i in range(self.MAX)]

    def get_hash(self, key):
        """remainder of the division of sum of ASCII char values by constant"""
        h = 0
        for char in key:
            h += ord(char)
        return h % self.MAX

    def __setitem__(self, key, val):
        """the __ is an overloading of python operators to utilize dict/array style calling methods"""
        h = self.get_hash(key)
        self.arr[h] = val

    def __getitem__(self, key):
        h = self.get_hash(key)
        return self.arr[h]
    
    def __delitem__(self, key):
        h = self.get_hash(key)
        self.arr[h] = None
    

```

```python
t = Hashtable()
t['Feb 16'] = 120
t['March 16'] = 40

t.arr
```
